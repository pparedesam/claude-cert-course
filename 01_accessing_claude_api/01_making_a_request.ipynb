````python
import os
import sys
sys.path.append('..')

from dotenv import load_dotenv
import anthropic

load_dotenv()  # Carga ANTHROPIC_API_KEY desde .env

# Verificar que la clave está configurada
api_key = os.environ.get('ANTHROPIC_API_KEY')
if api_key:
    print(f'API Key encontrada: {api_key[:12]}...')
else:
    print('Error: No se encontro ANTHROPIC_API_KEY')
````

## 3. Tu Primera Petición

````python
# Crear el cliente
client = anthropic.Anthropic()

# Hacer la primera peticion
message = client.messages.create(
    model="claude-sonnet-4-5-20250929",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "Hola Claude, puedes presentarte brevemente?"}
    ]
)

print(message.content[0].text)
````

## 4. Explorando la Estructura de la Respuesta

````python
# Explorar todos los campos de la respuesta
print("=== Objeto Message ===")
print(f"id:           {message.id}")
print(f"type:         {message.type}")
print(f"role:         {message.role}")
print(f"model:        {message.model}")
print(f"stop_reason:  {message.stop_reason}")
print(f"stop_sequence:{message.stop_sequence}")
print()
print("=== Uso de Tokens ===")
print(f"input_tokens:  {message.usage.input_tokens}")
print(f"output_tokens: {message.usage.output_tokens}")
print()
print("=== Content ===")
for i, block in enumerate(message.content):
    print(f"  block[{i}]: type={block.type}, text='{block.text[:80]}...'")
````

## 5. Parámetros Principales

| Parámetro | Descripción | Requerido |
|-----------|-------------|----------|
| `model` | Modelo a usar (ej. `claude-sonnet-4-5`) | Si |
| `max_tokens` | Máximo de tokens en la respuesta | Si |
| `messages` | Lista de mensajes (conversación) | Si |
| `system` | Instrucciones del sistema | No |
| `temperature` | Aleatoriedad (0.0-1.0) | No |
| `stream` | Activar streaming | No |


````python
# Comparar modelos disponibles
modelos = {
    "Opus":   "claude-opus-4-5-20251101",
    "Sonnet": "claude-sonnet-4-5-20250929",
    "Haiku":  "claude-haiku-4-5-20251001",
}

pregunta = "Cuanto es 2+2? Responde en una sola oracion."

for nombre, modelo_id in modelos.items():
    resp = client.messages.create(
        model=modelo_id,
        max_tokens=100,
        messages=[{"role": "user", "content": pregunta}]
    )
    print(f"[{nombre}] {resp.content[0].text.strip()}")
    print(f"  tokens: {resp.usage.input_tokens} in / {resp.usage.output_tokens} out")
    print()
````
